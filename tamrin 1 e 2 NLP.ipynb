```python
import re

text = '''در شکل 2-1 نرخ اطلاعات اصلی سیگنال (یا کنترل) در مراحل مختلف فرایند نیز آمده است. نرخ اطلاعات در متن اولیه و خام پیام، نسبتاً پایین است (حدود bps 50 (بیت بر ثانیه) مترادف با حدود 8 صدای مختلف در ثانیه، در شرایطی که هر صدا تقریباً 50 سمبل مجزا داشته باشد). بعد از تبدیل کد زبان، با اضافه¬شدن اطلاعات عروضی گفتار، نرخ اطلاعات تا bps 200 بالا می‌رود. در مراحل بعدی، ارائۀ اطلاعات در قالب سیگنال (یا کنترل) با سرعتی حدود bps 2000 به‌منظور کنترل عصبی تمامی اندام‌های درگیر در تولید گفتار و در حدود bps 50,000 – 30,000 در سطح سیگنال صوتی ادامه پیدا می‌کند.
روند تشخیص گفتار به‌صورت معکوس، الگوی فرایند تولید گفتار است؛ بنابراین، نرخ اطلاعات در غشای پایه در حدود bps 50,000 – 30,000 و در حین مرحلۀ هدایت عصبی در حدود bps 2000 است. بالاترین سطح پردازش در مغز، سیگنال‌های عصبی را به بازنمایی مجزا تبدیل می‌کند که سرانجام پیام با نرخ بیت پایین تولید می‌شود. 
احتراما پیرو هماهنگیهای انجام شده جهت بازدید دانش آموزان پایه نهم دبیرستان از آزمایشگاهها و کارگاههای دانشکده فنی و مهندسی دانشگاه الزهراء(س)، بدینوسیله درخواست خود جهت انجام بازدید دو گروه 20 نفری از دانش آموزان همراه با تعدادی از دبیران در روزهای چهارشنبه مورخ 22/1/97 و 29/1/97 ساعت 9 تا 12 را تقدیم می دارم.'''

# توکن بندی با علامت
words = re.findall(r"\w+\.\w+\,\w?|\d+\,+\d+|\w+|.\s|\.",text)
print(words)
```

```python
# تبدیل اعداد انگلیسی و حروف عربی به فارسی 

contractions_dict = { "أ": "ا", "إ":" ا", "آ": "ا", "ۀ": "ه", "ي": "ی", "ؤ": "و", "ة": "ه" ,"ئ": "ی" 
                     , "0": "۰" ,"1": "۱", "2": "۲", "3": "۳", "4": "۴", "5": "۵", "6": "۶", "7": "۷", "8": "۸", "9": "۹", ",": "", }

contractions_re = re.compile('(%s)'%'|'.join(contractions_dict.keys()))
def expand_contractions(s, contractions_dict = contractions_dict):
    def replace(match):
        return contractions_dict[match.group(0)]
    return contractions_re.sub(replace, s)
text = expand_contractions(text)


# توکن بندی بدون علامت

words = re.findall(r"\w+\‌\w+\w?|\d+\,+\d+|\d+\/\d+\/\d+|\w+",text)
print(words)
```

```python
class PersianStemmer:
    
    
    def get_base(self, word, suf):
        suflen = word.rfind(suf)
        base = word[:suflen]
        return base

# ایجاد تابع برای تشخیص حروف اخر
    
    def endsWith(self, stem):
        if re.findall(r'(\w+)s\b',stem):
            return True
        else:
            return False
        

# در مرحله اول حذف اعراب و تنوین از آخر کلمات

    def step1a(self, word):
        suffixes = ['ً' ,'ٌ' ,' ٍ' ,'َ' ,'ِ' ,'ُ' ,'ّ' ,'ء']
        for suffix in suffixes:
            if word.endswith(suffix):
                word = re.sub(suffix, '',word)
        return word
    
# حذف پیشوند ها 

    def step1b(self, word):

        if word.startswith('می'):
            word = re.sub( 'می', '',word)
        if word.startswith('نمی'):
            word = re.sub( 'نمی', '',word)
        if word.startswith('درمی'):
            word = re.sub( 'درمی', '',word)
        if word.startswith('برمی'):
            word = re.sub( 'برمی', '',word)
        if word.startswith('پیش'):
            word = re.sub( 'پیش', '',word)
        if word.startswith('غیر'):
            word = re.sub( 'غیر', '',word)
        if word.startswith('در'):
            word = re.sub( 'در', '',word)
        if word.startswith('به'):
            word = re.sub( 'به', '',word)
        if word.startswith('باز'):
            word = re.sub( 'باز', '',word)
        return word
    
# حذف پسوند ها از آخر کلمات

    def step2(self, word):
        
        suffixException = ["ها", "تر", "ترین", "ات", "اش", "های", "یی", "ی",'هایم', 'هایش', 'هایی', 'هایت', 'ان', 'گاه']
        for suffix in suffixException:
            if word.endswith(suffix):
                word = re.sub(suffix, '',word)
        return word
    
# stemmer فعل ها 
          
    def step3(self, word):
        if word.startswith('شده'):
            word = re.sub( 'شده', 'شد',word)
        if word.startswith('امده'):
            word = re.sub( 'امده', 'امد',word)
        if word.startswith('امدن'):
            word = re.sub( 'امدن', 'امد',word)
        if word.startswith('شدن'):
            word = re.sub( 'شدن', 'شد',word)
        if word.startswith('است'):
            word = re.sub( 'است', '',word)
        if word.startswith('داشته'):
            word = re.sub( 'داشته', 'داشت',word)
        if word.startswith('داشتن'):
            word = re.sub( 'داشتن', 'داشت',word)
        if word.startswith('بوده'):
            word = re.sub( 'بوده', 'بود',word)
        if word.startswith('بودن'):
            word = re.sub( 'بودن', 'بود',word)
        return word

# حذف حروف اضافه


    def step4(self, word):
        ezafe = ['از', 'در', 'به', 'با', 'بر', 'که', 'هر', 'تا', 'را', 'یا', 'و', 'برای', 'درباره',
                    'چون', 'مانند', 'مثل', 'بیرون', 'درون', 'زیر', 'بالا', 'روی', 'کنار', 'مقابل', 'مگر', 'توی', 'باز', 'پایین']
        for ezaf in ezafe:
            type(ezaf)
            if word == ezaf:
                word = re.sub(ezaf, '',word)
        return word           
    
```

```python
p = PersianStemmer()
for i in words:
    word = p.step1a(i)
    word = p.step1b(word)
    word = p.step2(word)
    word = p.step3(word)
    word = p.step4(word)
    if len(word) > 0:
        print(word,end= " ")
```
